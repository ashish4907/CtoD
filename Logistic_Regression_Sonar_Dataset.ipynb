```python
x = df.drop(columns=60,axis=0)
y = df[60]
```

# train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1281)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
pd.DataFrame(np.c_[y_test,y_pred],columns={'Original','Predicated'},index=y_test.index)
```

# StratifiedShuffleSplit

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=10,test_size=0.10,random_state=128)
for train_index,test_index in sss.split(x,y):
  x_train_sss = x.iloc[train_index]
  x_test_sss = x.iloc[test_index]
  y_train_sss = y.iloc[train_index]
  y_test_sss = y.iloc[test_index]
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_sss,y_train_sss)
y_pred_sss=model.predict(x_test_sss)
from sklearn.metrics import accuracy_score,confusion_matrix
cm_sss = confusion_matrix(y_test_sss,y_pred_sss)
print(cm_sss)
accuracy_score(y_test_sss,y_pred_sss)
```

```python
pd.DataFrame(np.c_[y_test_sss,y_pred_sss],columns={'Original','Predicated'},index=y_test_sss.index)
```
